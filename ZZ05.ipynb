### Zadanie 3:  Reprezentatywność próby losowej - część 2 [6 pkt]

Z populacji o rozkładzie N(172, 7) wylosowano ze zwracaniem 5000 prób losowych o liczebności 100. Na podstawie każdej z prób wyliczono jej wartość średnią, wariancję i odchylenie standardowe. Poniżej przedstawiono rozkłady wyliczonych parametrów:



Oceń w jaki sposób reprezentatywność próby losowej może wpływać na wyniki przeprowadzonych na jej podstawie analizy?

Reprezentatywność próby losowej ma zasadnicze znaczenie dla jakości wyników analizy. Jest to związane z możliwością uogólnienia wniosków na całą populację oraz precyzją uzyskanych rezultatów. Jeśli próba nie jest reprezentatywna, może to prowadzić do błędów statystycznych, czy też pojawienia się fałszywych zależności między zmiennymi. Nieprawidłowe próbkowanie może znacząco wpłynąć na interpretację wyników analizy i podejmowanie decyzji na ich podstawie. Staranne dobranie próby jest kluczowe dla uzyskania trafnych rezultatów. Wpływ eprezentatywności maleje, gdy zwiększa się ilość prób losowych. Wylosowanie 5000 prób sprawiły, że wartośći parametrów usśredniły się. Im więcej wykona się prób tym wynik statystycznie zbliża się do tego oczekiwanego, uśrednionego. Widoczne jest to np w przypadku kostki sześciennej, gdzie prawdopodobieństwo wyrzucenia poszczególnej wartości jest 1/6. W przypadku wykonania kilku prób jest duże prawdopodobieństwo, że nie otrzymamy zakładanych wyników i wyniki nie bedą w miarę równo rozłożone na każdą wartość kostki. Natomiast wraz ze wzrostem liczby prób ( np kilka tysięcy ), wartośći te się w miarę zrównają i liczba wyrzuconych danych oczek będzie podobna i zbliżona do siebie.
